Brief Column Explanations:

1) Year: the observation year.

2) Month: the observation month.

3) Temp: the difference in degrees Celsius between the average global temperature in that period and a reference value.

4) CO2, N2O and CH4 are expressed in ppmv

5) CFC.11 and CFC.12 are expressed in ppbv (parts per billion by volume). 

6) Aerosols: the mean stratospheric aerosol optical depth at 550 nm.

7) TSI: the total solar irradiance (TSI) in W/m2 (the rate at which the sun's energy is deposited per unit area).

8) MEI: multivariate El Nino Southern Oscillation index (MEI)
